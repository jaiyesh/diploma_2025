## Series::

Array+ Labels = Series

```python
import numpy as np
import pandas as pd
```

```python
por = np.array([0.25,0.32,0.16,0.12,0.10])
```

```python
por
```

```python
poro_sr = pd.Series(data = por, index = ["a","b","c","d","e"])
```

```python
poro_sr
```

```python
depth_wise_poro_series = pd.Series(data = por, index = [1000,1500,2000,2500,3000], name  = "Phi_vs_Depth")
```

```python
depth_wise_poro_series
```

```python
## plotting in pandas

depth_wise_poro_series.plot()
```
