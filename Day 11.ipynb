## Matplotlib: Plotting library from matlab

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Relative Permeability Plot

```python
sw = np.linspace(0,1,100)
```

```python
sw
```

```python
krw = sw**3

kro = (1-sw)**2
```

```python
krw 
```

```python
kro 
```

```python
# step1: krw vs sw plot

```

```python
plt.plot(sw,krw)
```

```python
#step 2: kro vs sw plot

plt.plot(sw,kro)
```

```python
# step 3: krw and kro plots together

plt.plot(sw,krw)
plt.plot(sw,kro)
```

```python
# step 4: Give x axis label

plt.plot(sw,krw)
plt.plot(sw,kro)

plt.xlabel("Water Saturation (fraction)")
```

```python
# step 5: Give y axis label

plt.plot(sw,krw)
plt.plot(sw,kro)

plt.xlabel("Water Saturation (fraction)")
plt.ylabel("Relative Permeability (fraction)")
```

```python
# step 6: grid

plt.plot(sw,krw)
plt.plot(sw,kro)

plt.xlabel("Water Saturation (fraction)")
plt.ylabel("Relative Permeability (fraction)")

plt.grid()
```

```python
#step 7 :  adding legend

plt.plot(sw,krw,label = "Water Relative permeability")
plt.plot(sw,kro, label = "Oil Relative Permeability")

plt.xlabel("Water Saturation (fraction)")
plt.ylabel("Relative Permeability (fraction)")


plt.legend()
plt.grid()

```

```python
# step 8: changing the size of the plot

plt.figure(figsize=(15,10))
plt.plot(sw,krw,label = "Water Relative permeability")
plt.plot(sw,kro, label = "Oil Relative Permeability")

plt.xlabel("Water Saturation (fraction)")
plt.ylabel("Relative Permeability (fraction)")


plt.legend()
plt.grid()
```

```python
# step 9 : change the line colour

plt.figure(figsize=(15,10))
plt.plot(sw,krw,label = "Water Relative permeability",color="#03dbfc")
plt.plot(sw,kro, label = "Oil Relative Permeability",color = "brown")

plt.xlabel("Water Saturation (fraction)")
plt.ylabel("Relative Permeability (fraction)")


plt.legend()
plt.grid()
```

```python
# step 10: changing the line thickness

plt.figure(figsize=(15,10))
plt.plot(sw,krw,label = "Water Relative permeability",color="#03dbfc",lw=4,ls = ":")
plt.plot(sw,kro, label = "Oil Relative Permeability",color = "brown",lw=2)

plt.xlabel("Water Saturation (fraction)")
plt.ylabel("Relative Permeability (fraction)")


plt.legend()
plt.grid()
```

```python
# step 11: adding title of the plot

plt.figure(figsize=(15,10))
plt.plot(sw,krw,label = "Water Relative permeability",color="#03dbfc",lw=4,ls = ":")
plt.plot(sw,kro, label = "Oil Relative Permeability",color = "brown",lw=2)

plt.xlabel("Water Saturation (fraction)")
plt.ylabel("Relative Permeability (fraction)")
plt.title("Relative Permeability Plot")

plt.legend()
plt.grid()
```

```python
# step 12 : style of the plot

plt.style.available
```

```python

plt.style.use("classic")

plt.figure(figsize=(15,10))
plt.plot(sw,krw,label = "Water Relative permeability",color="#03dbfc",lw=4,ls = ":")
plt.plot(sw,kro, label = "Oil Relative Permeability",color = "brown",lw=2)

plt.xlabel("Water Saturation (fraction)")
plt.ylabel("Relative Permeability (fraction)")
plt.title("Relative Permeability Plot")

plt.legend()
plt.grid(True)
```

```python
## step 13: saving the file


plt.style.use("classic")

plt.figure(figsize=(15,10))
plt.plot(sw,krw,label = "Water Relative permeability",color="#03dbfc",lw=4,ls = ":")
plt.plot(sw,kro, label = "Oil Relative Permeability",color = "brown",lw=2)

plt.xlabel("Water Saturation (fraction)")
plt.ylabel("Relative Permeability (fraction)")
plt.title("Relative Permeability Plot")

plt.legend()
plt.grid(True)

plt.savefig("RelPerm.png")
```
